# EIA API GET Request

```python
import pandas as pd
import datetime
import requests
import os
```

```python
eia_api_key = os.getenv('EIA_API_KEY')
```

```python
request = "https://api.eia.gov/v2/electricity/rto/region-data/data/?frequency=hourly&data[0]=value&facets[respondent][]=US48&facets[type][]=D&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000&api_key=" + eia_api_key 

d = requests.get(request).json()

d
```
